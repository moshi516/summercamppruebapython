> 🚨 **Punto de Control:** Asegúrate de subir `smash_dataset_clean.csv` a la carpeta de la izquierda antes de iniciar.

```python
# Importamos nuestras herramientas habituales
import pandas as pd
import sqlite3

# ¡NUEVA HERRAMIENTA! Plotly Express para gráficos interactivos
import plotly.express as px

# 1. Cargamos los datos rápidamente
df_torneo = pd.read_csv("smash_dataset_clean.csv", encoding="latin1")
conexion = sqlite3.connect("torneo_smash.db")
df_torneo.to_sql("combates", conexion, if_exists="replace", index=False)

# 2. Hacemos nuestra consulta SQL para traer la data
df_graficos = pd.read_sql_query("SELECT * FROM combates", conexion)
print("✅ Datos interactivos listos y cargados en memoria.")
```

## ⚖️ Misión 1: El Mapa Táctico Interactivo (Riesgo vs Recompensa)
Vamos a hacer un **Gráfico de Dispersión Interactivo**. La ventaja de Plotly es que, al pasar el cursor sobre un punto, ¡nos dirá exactamente qué jugador es y qué personaje usó sin tener que adivinar!

```python
# Creamos el gráfico interactivo con px.scatter
grafico_dispersion = px.scatter(
    df_graficos,
    x="Muertes",
    y="KOs",
    color="Personaje",       # Colores distintos por personaje
    size="KOs",              # El tamaño del punto depende de los KOs
    hover_data=["Nombre", "Escenario"], # ¡La magia! Datos extra al pasar el ratón
    title="⚖️ Análisis de Combate: Riesgo vs Recompensa",
    template="plotly_dark",  # Mantenemos el estilo Gamer
    size_max=20              # Tamaño máximo de los puntos
)

# Mostramos el gráfico
grafico_dispersion.show()
```

## 🍩 Misión 2: El Gráfico de Dona (La Meta del Torneo)
Vamos a descubrir qué personajes fueron los más elegidos en todo el campamento usando un gráfico circular especial con un agujero en el centro.

```python
# px.pie crea gráficos circulares, 'hole' le hace el agujero en medio
grafico_dona = px.pie(
    df_graficos,
    names="Personaje",
    hole=0.4,
    title="🔥 Personajes Más Populares del Torneo",
    template="plotly_dark"
)

# Resaltamos un poco los bordes para que se vea mejor
grafico_dona.update_traces(marker=dict(line=dict(color='#000000', width=2)))

grafico_dona.show()
```
