```python
import pandas as pd
import sqlite3
import plotly.express as px

# Re-create df_graficos within this cell since the previous cell was not executed
df_torneo = pd.read_csv("smash_dataset_clean.csv", encoding="latin1")
conexion = sqlite3.connect("torneo_smash.db")
df_torneo.to_sql("combates", conexion, if_exists="replace", index=False)
df_graficos = pd.read_sql_query("SELECT * FROM combates", conexion)

# Creamos el gráfico interactivo con px.scatter
grafico_dispersion = px.scatter(
    df_graficos,
    x="Edad",
    y="Muertes",
    color="Personaje",       # Colores distintos por personaje
    size="Muertes",              # El tamaño del punto depende de los KOs
    hover_data=["Nombre", "Escenario"], # ¡La magia! Datos extra al pasar el ratón
    title="⚖️ Análisis de Combate: Riesgo vs Recompensa",
    template="plotly_dark",  # Mantenemos el estilo Gamer
    size_max=20              # Tamaño máximo de los puntos
)

# Mostramos el gráfico
grafico_dispersion.show()
```

```python
# px.pie crea gráficos circulares, 'hole' le hace el agujero en medio
grafico_dona = px.pie(
    df_graficos,
    names="Edad",
    hole=0.4,
    title="🔥Edades con mas muertes en smash",
    template="plotly_dark"
)

# Resaltamos un poco los bordes para que se vea mejor
grafico_dona.update_traces(marker=dict(line=dict(color='#000000', width=2)))

grafico_dona.show()
```
