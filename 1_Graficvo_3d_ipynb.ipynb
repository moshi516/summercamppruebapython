```python
%%writefile mi_app.py
import streamlit as st
import pandas as pd
import plotly.express as px
st.set_page_config(layout="wide")
st.title("💎Mi mapa 3D de Animales")
st.write("Haz clic sostenido para girar la cueva y usa la rueda del ratón para hacer zoom")
datos = pd.DataFrame({
    "Animales": ["Perro", "Caballo", "Conejo", "Cerdito", "Pez", "Nutria"],
    "Capa_X": [10, 25, 5, 30, 15, 8],
    "Capa_Z": [15, 5, 20, 10, 25, 2],
    "Profundidad_Y": [50, 35, 15, -10, -30, -55], # El diamante está en lo más profundo
    "Experiencia": [10, 20, 30, 40, 50, 100],     # Esto hará que la burbuja sea más grande
    })
    # Creando el gráfico 3D interactivo
fig = px.scatter_3d(
    datos,
    x='Capa_X',
    y='Capa_Z',
    z='Profundidad_Y',
    color='Animales',
    size='Experiencia',
    hover_name='Animales',

    # Le damos los colores reales a los minerales
    color_discrete_map={
        "Perro": "#FBE29D",
        "Caballo": "#775537",
        "Conejo": "#F1F1F1",
        "Cerdito": "#FfD1D5",
        "Pez": "#FFA500",
        "Nutria": "#FBFFB2"
    }
)

# Apagamos las luces para que parezca una cueva oscura
fig.update_layout(
    template="plotly_dark",
    scene=dict(bgcolor="#111111"),
    margin=dict(l=0, r=0, b=0, t=0)
)

# Mostramos el gráfico en Streamlit
st.plotly_chart(fig, width='stretch')
```
